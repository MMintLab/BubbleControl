```python
import pandas as pd
import seaborn as sns
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
```

```python
datalegend_path = '/home/mmint/Desktop/drawing_evaluation/drawing_evaluation_DataLegend.csv'

df = pd.read_csv(datalegend_path)

df
```

```python
ax = sns.boxplot(x='ControllerMethod', y='Score', data=df, whis=3.)
```

```python
scores_learned_model = df[df.ControllerMethod == 'bubble_dynamics_pretrained_autoencoder_model_mppi']
mu = np.mean(scores_learned_model['Score'].values)
std = np.std(scores_learned_model['Score'].values)
print('MU:', mu)
print('STD:', std)
```

```python
scores_fixed_model = df[df.ControllerMethod == 'bubble_dynamics_fixed_model_mppi']
mu = np.mean(scores_fixed_model['Score'].values)
std = np.std(scores_fixed_model['Score'].values)
print('MU:', mu)
print('STD:', std)
```
